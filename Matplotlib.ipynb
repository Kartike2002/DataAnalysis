```python
import matplotlib.pyplot as plt
```

```python
#1 Create first chart
```

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x,y) #line chart
plt.show()  # display chart in notebook
```

```python
# 2 Customize chart
```

```python
plt.figure(figsize=(4,3)) #figure size

plt.plot(x, y, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.title("Kuch bhi ")
plt.xlabel("x-axis label ")
plt.ylabel("y-axis label")
plt.show()
```

```python
# 3 Advanced - multiple Lines & Legends
```

```python
x = [1,2,3,4,5]
y1 = [10,20,30,40,50]
y2 = [20,30,35,40,45]


plt.plot(x,y1,label='Sales 2024') #ploting y1 data
plt.plot(x,y2,label='Sales 2025') #ploting y2 data
plt.title("YOY Sales")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.legend()
plt.show()
```

```python
#4 Bar Chart
```

```python
x = ['A','B','C','D','E']
y=[10,20,25,35,45]

plt.bar(x,y)
plt.title('Bar chart Example')
plt.show()
```

```python
# Histogram
```

```python
# used for distribuation analysis

import random
data = [random.randint(1,10) for _ in range(100)]

plt.hist(data,bins=5)
plt.show()
```

```python
#6 Pie chart
```

```python
# used to show part-to-whole relationship/data

categories = ['A','B','C','D','E']
sales = [10,20,55,35,46]


plt.pie(sales,labels=categories,autopct= '%1.1f%%')
plt.title("Pie chart Example")
plt.show
```

```python
# 7 Scatter plot
```

```python
# used to find relationship btw variables

y1 = [10,20,30,40,50]
y2 = [15,25,32,28,34]

plt.scatter(y1,y2)
plt.title("Scatter plot")
plt.show()
```
