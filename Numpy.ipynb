```python
# Array Stacking and Splitting
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print(np.vstack((a,b)))  # row wise,Vertical stacking
print(np.hstack((a,b)))  # column wise, horizontal
```

```python
c = np.array([[1,2,3],[4,5,6]])
print(c)
```

```python
split = (np.hsplit(c,3))

for s in split:
  print(s)
```

```python
vsplit = (np.vsplit(c,2))

for s in vsplit:
  print(s)
```
