```python
# Concatenate and merge Dataframe (JOINS)
```

```python
df1 =pd.DataFrame({'ID':[1,2,3],'Name':['A','B','C']})
```

```python
df1
```

```python
df2 = pd.DataFrame({'ID':[1,2,3,4],'Score':[88,98,78,79]})
```

```python
df2
```

```python
pd.concat([df1,df2],axis=0) # row level / top on top
```

```python
pd.concat([df1,df2],axis=1) # cloumn level / side on side
```

```python
# merge = join
pd.merge(df1,df2,how='inner',on='ID')
```

```python
pd.merge(df1,df2,how='left',on='ID')
```

```python
pd.merge(df1,df2,how='right',on='ID')
```

```python
pd.merge(df1,df2,how='outer',on='ID')
```

```python
pd.merge(df1,df2,how='left',left_on='ID',right_on = 'ID')
```
